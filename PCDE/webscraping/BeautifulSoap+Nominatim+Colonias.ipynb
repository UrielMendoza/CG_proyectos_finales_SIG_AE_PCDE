```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent='cctmexico')

import time
start = time.time()

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

ColoniasDesabasto['direccion'] = ColoniasDesabasto['Texto'].apply(geocode)

end = time.time()
elapsed = end-start
print(str(elapsed)+"seconds")
```

```python
ColoniasDesabasto = ColoniasDesabasto.dropna()
ColoniasDesabasto.shape
```

```python
import time
start = time.time()

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

ColoniasDesabasto['coordenadas'] = ColoniasDesabasto['direccion'].apply(lambda x:(x.latitude, x.longitude))

end = time.time()
elapsed = end-start
print(str(elapsed)+"seconds")

```

```python
ColoniasDesabasto2 = ColoniasDesabasto.drop('Índice', axis=1)
ColoniasDesabasto2.head()
```

```python
latitud = []
longitud = []
for index, row in ColoniasDesabasto.iterrows():
  coordenadas = row['coordenadas']
  latitud.append(coordenadas[0])
  longitud.append(coordenadas[1])

ColoniasDesabasto['latitud'] = latitud
ColoniasDesabasto['longitud'] = longitud
```

```python
ColoniasDesabasto.head(15)
ColoniasDesabasto.to_csv('coloniasDesCDMX.csv', index=False, encoding='utf-8')
```

```python
import geopandas as gpd
from shapely.geometry import Point
import pandas as pd
```

```python
# Crear geometrías de puntos a partir de las coordenadas
geometry = [Point(xy) for xy in zip(ColoniasDesabasto2['longitud'], ColoniasDesabasto2['latitud'])]

# Crear un GeoDataFrame
gdf = gpd.GeoDataFrame(ColoniasDesabasto2, geometry=geometry)

# Definir el CRS (Sistema de Referencia de Coordenadas) para WGS 84
gdf.set_crs(epsg=4326, inplace=True)

# Transformar el CRS al sistema adecuado para CDMX si necesario (por ejemplo, a UTM zona 14N que es común para México)
gdf = gdf.to_crs(epsg=32614)

# Cambiar nombres de columnas para cumplir con el límite de longitud del shapefile
gdf.columns = [col[:10] for col in gdf.columns]

# Eliminar cualquier columna que pueda contener tipos de datos no primitivos
# Supongamos que 'location' podría ser una de esas columnas
if 'location' in gdf.columns:
    gdf.drop('location', axis=1, inplace=True)

# Asegurarse de que todos los datos son tipos básicos
for col in gdf.columns:
    if gdf[col].dtype == 'object':
        gdf[col] = gdf[col].astype(str)

# Guardar el GeoDataFrame como shapefile
gdf.to_file("colonias_cdmx.shp")
```
